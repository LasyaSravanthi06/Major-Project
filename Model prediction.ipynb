```python
import tensorflow  as tf
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input  #type: ignore
import numpy as np
import os
```

```python
model = tf.keras.models.load_model("Trained_Model.keras")
```

```python
model.summary()
```

```python
img_path = "test/CNV/CNV-28682-5.jpeg"
img = tf.keras.utils.load_img(img_path, target_size=(224, 224))
x = tf.keras.utils.img_to_array(img)
x = np.expand_dims(x,axis=0)
x = preprocess_input(x) # type: ignore
```

```python
x.shape
```

```python
preds = model.predict(x)
```

```python
preds
```

```python
class_name = ['CNV','DME','DRUSEN','NORMAL']
result_index = np.argmax(preds)
print("Model Prediction::",class_name[result_index])
```
